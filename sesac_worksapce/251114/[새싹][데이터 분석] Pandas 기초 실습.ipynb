````python
import pandas as pd
df = pd.read_csv('india_shares.csv') # 데이터셋 읽어오기

df['Date'] = pd.to_datetime(df['Date']) # 1. Date 컬럼을 날짜 타입으로 변경

df['Year'] = df['Date'].dt.year # 2. Date 컬럼의 dt 속성으로 연도(year) 속성 추출 후, 새 컬럼 생성
df['Month'] = df['Date'].dt.month # 2. Date 컬럼의 dt 속성으로 월 속성 추출 후 새 컬럼 생성

column = input() # 그룸화 할 컬럼의 이름 입력받기

grouped = df.groupby(column) # 입력 받은 컬럼으로 그룹화
df_mean = grouped.mean(numeric_only=True) # 숫자형 데이터에 대해 평균을 각 그룹마다 전체 컬럼에 대해

print(df_mean)
````
