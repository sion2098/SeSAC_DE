## 1. 과일 가격 출력하기


**문제 설명:**

과일 이름이 담긴 리스트가 주어집니다. for문을 사용하여 각 과일을 순서대로 출력하되, "N번째: 과일이름" 형식으로 출력하세요. (N은 1부터 시작)

**입력 예시:**

```
입력 없음
```

**출력 예시:**
```
1번째: 사과
2번째: 바나나
3번째: 포도
4번째: 오렌지
```

````python
fruits = ["사과", "바나나", "포도", "오렌지"]

i = 0
for fruit in fruits:
    i += 1
    print(f"{i}번째: {fruit}")
````


## 3. 카운트다운


**문제 설명:**

사용자로부터 숫자를 입력받아, 해당 숫자부터 0까지 카운트다운을 출력하세요. 0이 되면 "발사!"를 출력하고 종료합니다.

**입력 예시:**

```
5
```

**출력 예시:**
```
5
4
3
2
1
0
발사!
```

````python
start = int(input("숫자 입력: "))

while start != 0:
    print(start)
    start -= 1
    
    if start == 0:
        print(start)
        print("발사!")
````


## 5. 전화번호부 만들기


**문제 설명:**

3명의 이름과 전화번호를 입력받아 딕셔너리에 저장하세요. 그 후 사용자로부터 이름을 입력받아 해당 이름의 전화번호를 출력하세요. 이름이 없으면 "등록되지 않은 이름입니다."를 출력하세요.

**입력 예시:**

```
김철수
010-1234-5678
이영희
010-2345-6789
박민수
010-3456-7890

이영희
```

**출력 예시:**
```
이영희의 전화번호: 010-2345-6789
```

**입력 예시:**

```
김철수
010-1234-5678
이영희
010-2345-6789
박민수
010-3456-7890

홍길동
```

**출력 예시:**
```
등록되지 않은 이름입니다.
```

````python
phone_book = {}

book_name = [ ]
book_phone_number = [ ]

for i in range(3):
    name = input("이름: ")
    number = input("전화번호: ")
    book_name.append(name)
    book_phone_number.append(number)
# 리스트로 묶은거 출력해보기 print(book_name, book_phone_number)

phone_book = dict(zip(book_name, book_phone_number))
# 리스트 2개 딕셔너리로 잘 묶였는지 확인 print(phone_book)

key = input("누구?")
if key in phone_book.keys():
    print(f"{key}의 전화번호: {phone_book[key]}")
else:
    print("등록되지 않은 이름입니다")


#새로운 키 추가 방법으로 풀어보기

````


## 6. 재고 관리 시스템 (딕셔너리 + 반복문)


**문제 설명:**

상품명을 key, 재고 수량을 value로 하는 딕셔너리가 주어집니다.

- 전체 재고를 출력하세요.
- 재고가 10개 미만인 상품을 찾아 "부족" 표시와 함께 출력하세요.
- 전체 재고의 총 합계를 출력하세요.

**입력 예시:**

```
입력 없음
```

**출력 예시:**
```
=== 전체 재고 현황 ===
노트북: 15개
마우스: 8개
키보드: 12개
모니터: 5개
헤드셋: 20개

=== 재고 부족 상품 ===
마우스: 8개 (부족)
모니터: 5개 (부족)

총 재고: 60개
```

````python
inventory = {
    "노트북": 15,
    "마우스": 8,
    "키보드": 12,
    "모니터": 5,
    "헤드셋": 20
}

print("=== 전체 재고 현황 ===")
for x, y in inventory.items():
    print(x, ":" , y)
    
print("\n=== 재고 부족 상품 ===")
for x, y in inventory.items():
    if y < 10:
        print(f"{x} : {y}개 (부족)")

total = 0
for y in inventory.values():
    total += y
print(f"\n총 재고: {total} 개")
````


## 8. 업다운 게임

**문제 설명:**

정답 숫자를 미리 정해두고(예: 42), 사용자가 숫자를 입력할 때마다 "Up" 또는 "Down" 힌트를 주세요. 정답을 맞추면 "정답입니다! N번 만에 맞추셨습니다."를 출력하고 종료하세요.

Hint: while문 + break

**입력 예시:**

```
정답: 42

50
30
40
42
```

**출력 예시:**
```
Down!
Up!
Up!
정답입니다! 4번 만에 맞추셨습니다.
```

````python
# 원하는 정답 숫자를 하나 정하세요!
answer = int(input("원하는 정답 숫자는 :"))

# 사용자 숫자 입력 받기
user_num = int(input("당신의 입력: "))

count = 0
while True:
    if user_num > answer:
        print("Down!")
        user_num = int(input("당신의 입력: "))
        count += 1
    elif user_num < answer:
        print("Up!")
        user_num = int(input("당신의 입력: "))
        count += 1
    else:
        count += 1
        print(f"정답입니다! {count}번 만에 맞추셨습니다.")
        break

````


## 9. ATM 시뮬레이션

**문제 설명:**

초기 잔액 100,000원으로 시작하는 ATM 프로그램을 while문 무한 루프와 break를 활용하여 만드세요. 각 메뉴에 해당하는 번호를 입력 받고:

1. 현재 잔액 조회
2. 입금: 입금액을 추가로 입력받습니다. (잔액에 추가)
3. 출금: 출금액을 추가로 입력받습니다. (잔액에서 차감) (잔액 부족 시 에러 메시지)
4. 종료

"4"를 입력하면 프로그램을 종료합니다.

**입력 예시:**

```
1
2
50000
1
3
30000
1
4
```

**출력 예시:**
```
=== ATM ===
1. 잔액 조회
2. 입금
3. 출금
4. 종료
선택: 1
현재 잔액: 100000원

선택: 2
입금액: 50000
입금 완료! 현재 잔액: 150000원

선택: 1
현재 잔액: 150000원

선택: 3
출금액: 30000
출금 완료! 현재 잔액: 120000원

선택: 1
현재 잔액: 120000원

선택: 4
ATM을 종료합니다.
```

````python
print("=== ATM ===")
print("1. 잔액 조회", "\n2. 입금", "\n3. 출금", "\n4. 종료")
money = 100000

while True:
    num = input("번호 입력: ")

    if num == '1':
        print("선택:" , num)
        print("현재 잔액: ", money, "원")
        
    elif num == '2':
        add = int(input("입금액"))
        money += add
        print(f"선택: {num}")
        print(f"입금액: {add}")
        print(f"입금완료! 현재 잔액: {money}")
    elif num == '3':
        min = int(input("출금액"))
        money -= min
        print(f"선택: {num}")
        print(f"출금액: {min}")
        print(f"출금완료! 현재 잔액: {money}")
    elif num == '4':
        print("선택:", num)
        print("ATM을 종료합니다.")
        break
````
