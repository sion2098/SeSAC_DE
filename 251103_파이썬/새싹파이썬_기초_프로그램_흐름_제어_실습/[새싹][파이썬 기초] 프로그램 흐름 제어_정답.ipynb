**입력 예시:**

```
빨래하기
설거지하기
공부하기
운동하기
청소하기
3
```

**출력 예시:**
```
전체 할 일: ['빨래하기', '설거지하기', '공부하기', '운동하기', '청소하기']
'공부하기'를 완료했습니다!
남은 할 일: ['빨래하기', '설거지하기', '운동하기', '청소하기']

```

````python
# 빈 리스트를 만드세요
todo = []
# 할 일을 5개 입력 받고, 리스트에 추가하세요.
todo.append(input())
todo.append(input())
todo.append(input())
todo.append(input())
todo.append(input())

# 사용자에게 완료할 할 일의 번호를 입력받고, 번호 해당하는 할일을 삭제하세요.
todo_num = int(input())
todo.pop(todo_num - 1)
# 남은 할 일의 목록과 개수를 출력하세요.
print('남은 할 일:',todo)
print('남은 개수:', len(todo))
````


## 3. 주간 매출 분석


**문제 설명:**

일주일(월-일) 동안의 매출액이 리스트로 주어집니다. 평일(월금)과 주말(토~일)의 매출 합계를 각각 계산하여 출력하고, 평일 매출 합계와 주말 매출 합계의 차이를 출력하세요.

`sum(리스트)` 함수를 활용해보세요!
- 숫자 원소로 이루어진 리스트의 총 합계를 구하는 함수

**입력 예시:**

입력없음


**출력 예시:**
```
평일 매출 합계: 700만원
주말 매출 합계: 410만원
평일 매출 - 주말 매출: 290만원
```

````python
sales = [120, 150, 130, 140, 160, 200, 210]

# 매출 합계 계산과 출력
weekday = sum(sales[:5])
weekend = sum(sales[5:])
# 평일과 주말의 매출의 차이를 출력하세요.
print(f"평일 매출 합계: {weekday}만원")
print(f"주말 매출 합계: {weekend}만원")
print(f"평일 매출 - 주말 매출: {weekday - weekend}만원")
````


## 4. 재고 관리 시스템


**문제 설명:**

창고의 상품 재고 리스트가 주어집니다. 다음 작업을 순서대로 수행하세요:

1. "노트북" 2개를 추가
2. 첫 번째로 나오는 "마우스"를 출고(삭제)
3. 재고를 알파벳 순으로 정렬
4. "키보드"의 재고가 몇 개인지 출력
5. 최종 재고 리스트 출력

**입력 예시:**
```
입력없음
```

**출력 예시:**
```
키보드 재고: 3개
최종 재고: ['노트북', '노트북', '마우스', '모니터', '키보드', '키보드', '키보드']



```

````python
inventory = ["마우스", "키보드", "모니터", "마우스", "키보드", "키보드"]

inventory.append("노트북")
inventory.append("노트북")

inventory.remove("마우스")

inventory.sort()

print('키보드 재고:',inventory.count("키보드"))
print('최종 재고:',inventory)
````

## 4-2. 중복 제거 프로그램
주어진 리스트에서 중복된 숫자를 제거하고 오름차순으로 정렬된 리스트를 만드세요.

- 리스트를 (눈으로)확인해서 count()를 사용해 중복된 숫자와 개수 출력
- 중복되는 값은 remove()로 제거(= 각 값이 한개만 남도록)
- sort()로 오름차순 정렬
- 최종 리스트 출력

**출력 예시**
```
중복 값: 5, 개수: 3
중복 값: 8, 개수: 2
중복 값: 2, 개수: 2
[1, 2, 3, 5, 8, 9]
```

````python
numbers = [5, 2, 8, 2, 9, 5, 3, 8, 1, 5]
print(f"중복 값: 5, 개수: {numbers.count(5)}")
print(f"중복 값: 8, 개수: {numbers.count(8)}")
print(f"중복 값: 2, 개수: {numbers.count(2)}")

numbers.remove(5)
numbers.remove(5)
numbers.remove(8)
numbers.remove(2)
numbers.sort()
print(numbers)
````

**문제 설명:**

각 학생의 평균을 계산하고, 각 학생의 이름과 평균을 출력하세요.

`sum(리스트)` 함수를 활용하세요!
- 숫자 원소로 이루어진 리스트의 총 합계를 구하는 함수

**입력 예시:**

```

```

```
김철수 84.33333333333333 
이영희 91.66666666666667
박민수 69.0
```

````python
students = [
    ["김철수", 85, 90, 78],
    ["이영희", 92, 88, 95],
    ["박민수", 70, 65, 72]
]

s1_total = sum(students[0][1:])
s2_total = sum(students[1][1:])
s3_total = sum(students[2][1:])

s1_avg = s1_total / 3
s2_avg = s2_total / 3
s3_avg = s3_total / 3


print(students[0][0], s1_avg)
print(students[1][0], s2_avg)
print(students[2][0], s3_avg)
````


## 8. 로그인 시스템


**문제 설명:**

등록된 아이디와 비밀번호가 주어집니다. 사용자로부터 아이디와 비밀번호를 입력받아:

- 둘 다 일치: "로그인 성공"
- 아이디만 틀림: "존재하지 않는 아이디입니다"
- 비밀번호만 틀림: "비밀번호가 틀렸습니다"
- 둘 다 틀림: "아이디와 비밀번호를 확인하세요"

위 조건에 따라 출력하세요.

**입력 예시:**

```
user123
pass456
```


**출력 예시:**

```
로그인 성공

```

````python
correct_id = "user123"
correct_pw = "pass456"

id = input()
password = input()

if correct_id == id and correct_pw == password:
    print("로그인 성공")
elif correct_id != id and correct_pw == password:
    print('존재하지 않는 아이디입니다.')
elif correct_id == id and correct_pw != password:
    print('비밀번호가 틀렸습니다.')
else:
    print('아이디와 비밀번호를 확인하세요.') 
````


## 9. 택배 요금 계산기


**문제 설명:**

물건의 무게(kg)와 배송 지역(서울/지방/제주)을 입력받아 배송비를 계산하세요.

- 서울: 3kg 이하 3000원, 3kg 초과 시 1kg당 1000원 추가
- 지방: 3kg 이하 4000원, 3kg 초과 시 1kg당 1500원 추가
- 제주: 3kg 이하 5000원, 3kg 초과 시 1kg당 2000원 추가

**입력 예시:**

```
5
지방
```


**출력 예시:**

```
배송 지역: 지방
무게: 5kg
배송비: 7000원
```
(설명: 기본 4000원에 초과 2kg에 의한 추가 3000원)


````python
weight = float(input("물건의 무게(kg)를 입력하세요: "))
region = input("배송 지역을 입력하세요 (서울/지방/제주): ")

# 배송비 계산
if region == "서울":
    base_fee = 3000
    extra_rate = 1000
elif region == "지방":
    base_fee = 4000
    extra_rate = 1500
elif region == "제주":
    base_fee = 5000
    extra_rate = 2000

if weight <= 3:
    fee = base_fee
else:
    fee = base_fee + (weight - 3) * extra_rate

print(f"\n배송 지역: {region}")
print(f"무게: {weight}kg")
print(f"배송비: {fee}원")
````


## 10. 혈압 판정 시스템


**문제 설명:**

사용자로부터 수축기 혈압과 이완기 혈압을 입력받아 상태를 판정하세요.

- 수축기 120 미만 AND 이완기 80 미만: "정상"
- 수축기 120~129 AND 이완기 80 미만: "주의"
- 수축기 130~139 OR 이완기 80~89: "고혈압 전단계"
- 수축기 140 이상 OR 이완기 90 이상: "고혈압"

**입력 예시:**

```
135
85
```


**출력 예시:**

```
수축기: 135mmHg 이완기: 85mmHg
판정: 고혈압 전단계

```


````python
systolic = int(input("수축기 혈압(mmHg)을 입력하세요: "))
diastolic = int(input("이완기 혈압(mmHg)을 입력하세요: "))

# 판정 조건
if systolic < 120 and diastolic < 80:
    status = "정상"
elif 120 <= systolic <= 129 and diastolic < 80:
    status = "주의"
elif (130 <= systolic <= 139) or (80 <= diastolic <= 89):
    status = "고혈압 전단계"
elif systolic >= 140 or diastolic >= 90:
    status = "고혈압"

# 결과 출력
print(f"\n수축기: {systolic}mmHg 이완기: {diastolic}mmHg")
print(f"판정: {status}")
````
