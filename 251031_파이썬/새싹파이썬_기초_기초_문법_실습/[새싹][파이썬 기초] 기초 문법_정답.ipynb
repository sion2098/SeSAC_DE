
## 3. 자진모리장단

**문제 설명:**

**입력 예시:**

```
입력없음
```

**출력 예시:**
```
덩덕쿵덕

덩덕쿵덕쿵덕쿵덕덩덕쿵덕쿵덕쿵덕

```

````python
# 변수 하나를 정의하고 + 연산자를 활용하여 "덩덕쿵덕"을 완성하시오.
connect_str = "덩덕" + "쿵덕"
# 변수 하나를 정의하고 * 연산자를 활용하여 "쿵덕쿵덕"을 완성하시오.
iterate_str = "쿵덕" * 2
# 위 두 변수를 활용하여 "덩덕쿵덕쿵덕쿵덕덩덕쿵덕쿵덕쿵덕"을 완성하시오.
jajinmori = (connect_str + iterate_str) * 2
# 완성한 세 변수를 출력하시오.
print(connect_str)
print(iterate_str)
print(jajinmori)
````


## 14. 참/거짓 판별 명제만들기 (4) 윤년 판별하기

**문제 설명:**

연도를 입력받았을 때, 해당 연도가 윤년이면 True, 그렇지 않으면 False가 되는 명제를 완성하시오.

윤년의 조건
- 윤년은 4의 배수이면서, 100의 배수가 아닐 때 또는 400의 배수이다. 

예시
- 2012년은 4의 배수이면서 100의 배수가 아니라서 윤년.
- 1900년은 100의 배수이고 400의 배수가 아니기 때문에 윤년이 아님
- 2000년은 400의 배수이기 때문에 윤년이다.


**입력 예시 1:**
```
2012
```
**출력 예시 1:**
```
True

```

**입력 예시 2:**
```
1900
```
**출력 예시 2:**
```
False

```



````python
# 연도를 입력받으시오.
year = int(input())

# 윤년을 판별하는 명제를 완성하시오.
print((year % 4 == 0 and year % 100 != 0) or (year % 400 == 0))
````
